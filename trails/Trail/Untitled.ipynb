```python
def get_pdf_data(path):
    with open(path,'rb') as file:
        text = pdf.PdfReader(file)
        texts = []
        for i in range(len(text.pages)):
            page = text.pages[i]
            get_text = page.extract_text()
            texts.append(get_text)
            return(''.join(texts))
        
```

```python
get_pdf_data('Food Calories List.pdf')
```
