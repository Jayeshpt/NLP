```python
import csv

data = [['Name','Age','Gender'],
        ['Jayesh',"20",'male'],
        ['sooraj','30','male'],
        ['hisham','30','male']]

with open('newout.csv','w',newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```
